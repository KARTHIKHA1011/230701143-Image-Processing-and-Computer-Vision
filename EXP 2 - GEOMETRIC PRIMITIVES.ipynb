```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------- 1. Scaling --------
img1 = cv2.imread("messi5.jpg")
resized = cv2.resize(img1, None, fx=2, fy=2)

# -------- 2. Translation --------
rows, cols = img1.shape[:2]
M = np.float32([[1, 0, 100], [0, 1, 50]])
translated = cv2.warpAffine(img1, M, (cols, rows))

# -------- 3. Rotation --------
M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)
rotated = cv2.warpAffine(img1, M, (cols, rows))

# -------- 4. Affine --------
img2 = cv2.imread("drawing.png")
rows2, cols2 = img2.shape[:2]

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1, pts2)
affine = cv2.warpAffine(img2, M, (cols2, rows2))

# -------- 5. Perspective --------
img3 = cv2.imread("sudoku.png")

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1, pts2)
perspective = cv2.warpPerspective(img3, M, (300,300))

# -------- Display --------
plt.figure(figsize=(10,7))

images = [resized, translated, rotated, affine, perspective]
titles = ["Scaling", "Translation", "Rotation", "Affine", "Perspective"]

for i in range(5):
    plt.subplot(2,3,i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
